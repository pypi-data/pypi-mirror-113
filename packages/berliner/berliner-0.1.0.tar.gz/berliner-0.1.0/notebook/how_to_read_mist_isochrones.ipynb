```python
from berliner import mist
isocs_ps1 = mist.read_isochrones_ptn("/hydrogen/mist/1.2/isochrones/MIST_v1.2_vvcrit0.4_PanSTARRS/MIST_v1.2_*_vvcrit0.4_PanSTARRS.iso.cmd")
```

```python
isoc_mhini = np.array([_.meta["MH"] for _ in isocs_ps1])
isoc_lgage = np.array([_.meta["lgage"] for _ in isocs_ps1])

# plot age-metallicity
plot(isoc_lgage, isoc_mhini, 'x')
xlabel("log(Age/yr)")
ylabel("[M/H] (dex)");
```

## Isochrone Grid

```python
from berliner.utils import IsochroneGrid
ig = IsochroneGrid(isoc_lgage, isoc_mhini, isocs_ps1, model="mist")
```

```python
isoc0 = ig.get_a_isoc(9,0)
isoc0
```

```python
for _lgage in np.arange(6,10,0.05):
    isoc0 = ig.get_a_isoc(_lgage,0)
    plot(isoc0["log_Teff"],isoc0["log_g"],)
```

```python
for _lgage in np.arange(6,10,0.05):
    isoc0 = ig.get_a_isoc(_lgage,0)
    plot(isoc0["EEP"],isoc0["log_g"],)
```
